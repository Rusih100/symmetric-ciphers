```python
INPUT_FILE_PATH = r"D:\Pycharm-project\crypto_labs\data\lab2\2B.txt"

with open(INPUT_FILE_PATH, encoding="utf-8") as file:
    raw_text = file.read().strip().lower().split()
    text = "".join(raw_text)
```

```python
text
```

### Частоты всех триграмм

```python
treegrams = list(ngram_frequencies(text, 3).items())
treegrams
```

### Сортируем триграммы по частотам

```python
treegrams.sort(key=lambda x: x[1], reverse=True)
dot = treegrams[0][0]
comma = treegrams[1][0]

if dot[0] != comma[2]:
    dot, comma = comma, dot

print(dot, comma)
```

### Задаем алфавит замены

```python
alphabet: dict[str, str] = {
    dot[0]: "т",
    dot[1]: "ч",
    dot[2]: "к",
    comma[0]: "з",
    comma[1]: "п",
}
alphabet
```

### Функция просмотра зашифрованного текста

```python
def show_encrypt_text(text: str, alphabet: dict[str, str]) -> None:
    result = ""
    for char in text:
        if char in alphabet:
            result += alphabet[char]
        else:
            result += "_"
    print(result)
```
