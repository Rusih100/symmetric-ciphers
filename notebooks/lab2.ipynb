```python
from src.text_statistics import ngram_frequencies
from collections import Counter
import re
```

```python
INPUT_FILE_PATH = r"D:\Pycharm-project\crypto_labs\data\lab2\2B.txt"

with open(INPUT_FILE_PATH, encoding="utf-8") as file:
    raw_text = file.read().strip().lower().split()
    text = "".join(raw_text)
```

```python
text
```

### Частоты всех триграмм

```python
treegrams = list(ngram_frequencies(text, 3).items())
treegrams
```

### Сортируем триграммы

```python
treegrams = sorted(treegrams, key=lambda x: x[1], reverse=True)
dot = treegrams[0][0]
comma = treegrams[1][0]

if dot[0] != comma[2]:
    dot, comma = comma, dot

print(dot, comma)
```

### Находим все слова разделенные точкой и запятой

```python
words = re.split(f"{dot}|{comma}", text)
words
```

### Задаем алфавит замены

```python
alphabet_decrypt: dict[str, str] = {
    dot[0]: "т",
    dot[1]: "ч",
    dot[2]: "к",
    comma[0]: "з",
    comma[1]: "п",
}
alphabet_encrypt = {v: k for k, v in alphabet_decrypt.items()}
```

### Ищем "Штирилиц"

```python
found: list[str] = []

for i in range(len(text) - 7 + 1):
    word = text[i:i+ 7]
    
    if not word[1] == alphabet_encrypt['т']:
        continue
    if not word[2] == word[5]:
        continue        
    if not len(set(word)) == 6:
        continue
    
    flag = False
    for j in (0, 2, 3, 4, 5, 6):
        if word[j] in alphabet_decrypt:
            flag = True
            break
            
    if flag:
        continue
        
    found.append(word)

found_word = Counter(found).most_common(1)[0][0]
found_word
```

### Обновляем алфавиты

```python
check_word = "штирлиц"

for i, char in enumerate(found_word):
    if char not in alphabet_decrypt:
        alphabet_decrypt[char] = check_word[i]
        alphabet_encrypt[check_word[i]] = char
    
alphabet_decrypt
```

### Ищем "Мюллер"

```python
found: list[str] = []

for i in range(len(text) - 6 + 1):
    word = text[i:i+ 6]
    
    if not word[2] == word[3] and word[2] == alphabet_encrypt["л"]:
        continue
    if not word[5] == alphabet_encrypt["р"]:
        continue
    if not len(set(word)) == 5:
        continue
    
    flag = False
    for j in (0, 1, 4):
        if word[j] in alphabet_decrypt:
            flag = True
            break
            
    if flag:
        continue
    
    found.append(word)

found_word = Counter(found).most_common(1)[0][0]
found_word
```
